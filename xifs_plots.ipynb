```python
import xarray as xr
import matplotlib.pyplot as plt
import xifs
import cartopy.crs as ccrs
```

```python
file='ECE3_1m_19790101-20191231_regular_sfc.nc'
```

```python
analysis_list=["glomean_crf","seasmean_t2m","seasmean_sp","es",'tclw','tciw','sp','tcwv','lsp','cp','sf','sshf','slhf','msl','2t','2d','ssrd','strd','ssr','str','tsr','ttr','e','tsrc','ttrc','ssrc','strc','tisr','tp','skt']
```

```python
result=xifs.analysis(analysis_list,file)
```

```python
list_2D=["glomean_crf","es",'tclw','tciw','sp','tcwv','lsp','cp','sf','sshf','slhf','msl','2t','2d','ssrd','strd','ssr','str','tsr','ttr','e','tsrc','ttrc','ssrc','strc','tisr','tp','skt']
seasons=['DJF', 'MAM', 'JJA', 'SON']
```

```python
for l in list_2D:
    plt.figure(figsize=(15,5))
    result[l].plot();
```

```python
for s in seasons:
    plt.figure(figsize=(10,5))
    p=result['seasmean_t2m'].sel(season=s).plot(cmap='Spectral_r',
        subplot_kws=dict(projection=ccrs.Mollweide(0), facecolor="gray"),
        transform=ccrs.PlateCarree(),)
    p.axes.coastlines()
```

```python
for s in seasons:
    plt.figure(figsize=(10,5))
    p=result['seasmean_sp'].sel(season=s).plot(cmap='Spectral_r',
        subplot_kws=dict(projection=ccrs.Mollweide(0), facecolor="gray"),
        transform=ccrs.PlateCarree(),)
    p.axes.coastlines()
```
